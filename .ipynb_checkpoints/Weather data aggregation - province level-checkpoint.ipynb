```python
from rasterstats import zonal_stats
import geopandas as gpd
import pandas as pd
import os
import glob
import numpy as np
from tqdm import tqdm
```

```python
#input/output directories and files
raster_file_dir = 'raw_data'
output_dir = 'data_aggregated_per_province'
shapefile = 'shapefiles/phl_admbnda_adm2_psa_namria_20180130.shp'
#read shape
gdf_adm = gpd.read_file(shapefile)
```

```python
gdf_adm.head()
```

```python
#get admin names
adm_levels = gdf_adm['ADM2_EN'].tolist()
#create dates
dates = pd.date_range(start='1/1/2012', end='12/1/2019', freq='MS')
#create empty df with multiindex admin/date
df_final = pd.DataFrame(index=pd.MultiIndex.from_product([adm_levels, dates],
                                                      names=['adm_level', 'date']))
```

```python
os.listdir(raster_file_dir)
```

```python
#for each feature_folder 
for dir_col in os.listdir(raster_file_dir):
    print(dir_col)
    df_final[dir_col] = np.nan
    #create temp df with same index
    df_partial = pd.DataFrame(index=pd.MultiIndex.from_product([adm_levels, dates],
                                                      names=['adm_level', 'date']))
    df_partial[dir_col] = np.nan
    
    #for each month
    for date_month in tqdm(dates):
        #create string from the date
        date = date_month.strftime('%Y-%m-%d')
        raster_file_search = raster_file_dir+'/'+dir_col+'/*_'+date+'_*/*.tif'
        raster_file_matched = glob.glob(raster_file_search)
        for raster_file in raster_file_matched:
            stats = zonal_stats(shapefile,
                                raster_file,
                                stats="median")
            for idx, stat_region in enumerate(stats):
                for key, value in stat_region.items():
                    df_final.at[(adm_levels[idx], date), dir_col] = value
                    df_partial.at[(adm_levels[idx], date), dir_col] = value
    df_partial.to_csv(output_dir+'/'+dir_col+'_adm2.csv')

df_final.to_csv(output_dir+'/aggregated_adm2_median.csv')
```
